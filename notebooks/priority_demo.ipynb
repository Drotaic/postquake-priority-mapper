# 🚨 Postquake Priority Map Demo
Simulates post-disaster UAV priority scoring for search & rescue routing.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scripts.priority_map import compute_priority

# Example data
grid = 5
data = {
    'damage_level': np.random.randint(0, 4, grid * grid),
    'survivors_detected': np.random.randint(0, 3, grid * grid),
    'road_blocked': np.random.randint(0, 2, grid * grid),
    'distance_to_hospital': np.random.uniform(0, 50, grid * grid)
}
df = pd.DataFrame(data)
df['priority_score'] = df.apply(compute_priority, axis=1)
heatmap = df['priority_score'].values.reshape((grid, grid))
plt.imshow(heatmap, cmap='hot', interpolation='nearest')
plt.title("Priority Heatmap")
plt.colorbar(label="Score")
plt.show()
```
